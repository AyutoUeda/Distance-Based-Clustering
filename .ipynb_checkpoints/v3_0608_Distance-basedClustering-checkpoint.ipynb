```python
coordinate_data = []
t = 10000
# 何秒目をクラスタリングするか
for i in range(0, 120, 2):
  x = df.iloc[t][i]
  y = df.iloc[t][i+1]
  coordinate_data.append([int(x), int(y)])

# list --> array
X = np.array(coordinate_data)

def plot_coordinate(X, str, num1, num2):    
    for i in range(num1, num2):
        distance_threshold = i
        model = AgglomerativeClustering(
            n_clusters=None, linkage=str, compute_full_tree=True, distance_threshold=distance_threshold
        )

        model.fit(X)

        # 位置とクラスタ番号の描画
        fig = plt.figure(figsize=(6,6))
        plt.scatter(X[:,0], X[:,1], c=model.labels_) #cmap="viridis")

        labels = model.labels_

        clusters_size = np.bincount(labels) # <-- ラベル別のクラスタサイズ

        n = sum(x>1 for x in clusters_size) # <--クラスタサイズが1より大きいものの数

        plt.title("distance based (%s) \n distance = %d, n_clusters=%d" %(str, distance_threshold, n))

        # ラベルのプロット
        for i in range(60):
    #         plt.annotate(labels[i], (X[i][0], X[i][1]))

            if clusters_size[labels[i]] > 1:
                plt.annotate(labels[i], (X[i][0], X[i][1]))


        plt.show()
```

## 関数作成

```python
def hierachical_clustering(num1, num2):
    '''
    num1 --> 分析する秒数（行数）
    num2 --> クラスタ間の距離
    '''
    coordinate_data = []
    t = num1
    # 何秒目をクラスタリングするか
    for i in range(0, 120, 2):
      x = df.iloc[t][i]
      y = df.iloc[t][i+1]
      coordinate_data.append([int(x), int(y)])

    # list --> array
    X = np.array(coordinate_data)

    distance_threshold = num2
    model = AgglomerativeClustering(
        n_clusters=None, linkage="single", compute_full_tree=True, distance_threshold=distance_threshold
    )

    model.fit(X)
    
    return model.n_clusters_, model.labels_, coordinate_data, model

'''
戻り値
coordinate_data --> 指定した行数における各点の座標 
'''
```

## クラスタサイズが1のものを除くプログラム

## クラスタ数と距離（クラスタサイズが1より大きいもののみ）

```python
'''
・クラスタ間の距離を変化させる
・処理する時間は固定
'''
time = 10000
clusters1 = [] # クラスタ数
dis = [s for s in np.arange(30, 50, 0.5)]
for i in np.arange(30, 50, 0.5):
    labels = hierachical_clustering(time,i)[1] # 各点のラベル

    clusters_size = np.bincount(labels) # ラベル別のクラスタサイズ

    n = sum(x>1 for x in clusters_size) # クラスタサイズが1より大きいもののみカウント
    clusters1.append(n)

print(clusters1)


X = dis
Y = clusters1
plt.plot(X, Y)
plt.xlabel("distance_threshold")
plt.ylabel("n_clusters")
plt.title("Relationship between distance_threshold (0.5steps) and n_clusters \n time=%d" %time)
plt.show()
```

ある一点におけるクラスタ数と距離の閾値の関係（クラスタサイズが1のものを除く）

## 時間ごとのクラスタ数の変化(クラスタサイズが1より大きいもの）

```python
%%time
'''
・時間ごとのクラスタ数の表示
'''
start = 5000
end = 10000
clusters2 = []
selected_time = [s for s in range(start, end)]

for i in range(30, 36):
    c = []
    for j in range(start, end):
        labels = hierachical_clustering(j,i)[1] # 各点のラベル

        clusters_size = np.bincount(labels) # ラベル別のクラスタサイズ

        n = sum(x>1 for x in clusters_size) # クラスタサイズが1より大きいもののみカウント
        c.append(n)
    clusters2.append(c)
```

### **memo**  
5000行目から5100行目までにおける閾値を5通り試すのに7.39s  
5000行目から10000行目まで--> 6min 6s

```python
np.mean(clusters2[0])
```

```python
distance = [s for s in range(30, 35)]
for i in range(len(distance)):
    mean = np.mean(clusters2[i]) # クラスタサイズの平均値

    plt.figure(figsize=(15,4))
    plt.plot(selected_time, clusters2[i])
    plt.plot([start, end],[mean, mean], "red", linestyle='dashed', label="mean") # 平均値を示す補助線
    plt.xlabel("time")
    plt.ylabel("clusters")
    plt.ylim((5,23))
    plt.xlim((start, end))
    plt.title("Relationship between time and n_clusters \n distance_threshold=%d \n min=%d , max=%d , mean=%1.2f" 
              %(distance[i], min(clusters2[i]), max(clusters2[i]), mean))
    plt.legend()
    plt.show()
```

memo  
距離の閾値が上がるほどクラスタ数の平均値が増える

## クラスタサイズの時間変化（最大値）

```python
'''
・時間ごとのクラスタ数の表示
'''
start1 = 5000
end1 = 10000
clusters3 = [] # クラスタサイズの最大値
clusters4 = [] # クラスタサイズの最小値
selected_time1 = [s for s in range(start1, end1)]

for i in range(30, 35):
    c_max = []
    c_min = []
    for j in range(start1, end1):
        labels = hierachical_clustering(j,i)[1] # 各点のラベル

        clusters_size = np.bincount(labels) # ラベル別のクラスタサイズ
        except_size1 = [x for x in clusters_size if x > 1]

        maxnum = max(except_size1) # クラスタサイズが最大のものを抽出
        c_max.append(maxnum)
        
        minimum = min(except_size1)
        c_min.append(minimum)
    clusters3.append(c_max)
    clusters4.append(c_min)
```

```python
# 描画
distance = [s for s in range(30, 35)]
for i in range(len(distance)):
    mean = np.mean(clusters3[i]) # 最大クラスタサイズの平均値

    plt.figure(figsize=(15,4))
    plt.plot(selected_time1, clusters3[i], label="maximum")
    plt.plot(selected_time1, clusters4[i], c="black", label="minimum")
    plt.plot([start1, end1],[mean, mean], "red", linestyle='dashed', label="mean of maximum") # 平均値を示す補助線
    plt.xlabel("time")
    plt.ylabel("max cluster size")
    plt.ylim((0,15))
    plt.xlim((start1, end1))
    plt.title("Relationship between time, maximum clustersize and minimum clustersize \n distance_threshold=%d \n mean of maximum=%1.2f" 
              %(distance[i], mean))
    plt.legend()
    plt.show()
```

## 樹形図

```python
# 樹形図　描画用関数
def plot_dendrogram(model, **kwargs):
   # Create linkage matrix and then plot the dendrogram

   # create the counts of samples under each node
   counts = np.zeros(model.children_.shape[0])
   n_samples = len(model.labels_)
   for i, merge in enumerate(model.children_):
      current_count = 0
      for child_idx in merge:
         if child_idx < n_samples:
            current_count += 1 # leaf node
         else:
            current_count += counts[child_idx - n_samples]
      counts[i] = current_count

   linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)

   # Plot the corresponding dendrogram
   dendrogram(linkage_matrix, leaf_font_size=12, **kwargs)
```

```python
# 樹形図の描画
plt.figure(figsize=(20,10))
plt.title('Euclidean, linkage=single')
# plot the top three levels of the dendrogram
plot_dendrogram(hierachical_clustering(1000,30)[3], truncate_mode='level', p=10)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.ylabel("Distance")
ax = fig.add_subplot(111)

# plt.plot([0, 40],[40, 40], "red", linestyle='dashed') 
plt.show()
```
